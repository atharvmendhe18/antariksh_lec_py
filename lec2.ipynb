```python
def calc(a,b,c):
  if c==1:
    return a+b
  elif c==2:
    return a-b
  elif c==3:
    return a*b
  elif c==4:
    return a/b
  else:
    print("invalid")
r = calc(6,4,1)
print(r)
```
